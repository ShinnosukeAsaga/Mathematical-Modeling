```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 与えられたデータ
X = np.array([1, 2, 3])
Y = np.array([2, 3, 5])

# 最小二乗法で回帰直線の係数を求める
a, b = np.polyfit(X, Y, 1)

# 回帰直線を描画するためのY値
Y_pred = a * X + b

# グラフの描画
plt.figure(figsize=(6, 4))
plt.scatter(X, Y, color='blue', label='Ture_data')
plt.plot(X, Y_pred, color='red', label=f'Regression line: y = {a:.2f}x + {b:.2f}')
plt.title('Ordinary Least Squares regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 結果の表示
a, b
```
